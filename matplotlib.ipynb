```python
import numpy as n
import matplotlib.pyplot as plt
x=n.array([1,2,6,8])
y=n.array([1,2,6,8])
plt.plot(x,y,ms=10,marker='s',linestyle=':',mfc='g',mec='k')
plt.title('graph')
plt.plot(x,y)
plt.show()
```

```python
import numpy as n
import matplotlib.pyplot as plt
x=n.array([2,5,7,9])
y=[0,0,0.2,0]
a=["linux","mac","ubuntu","windows",]
plt.pie(x,labels=a,explode=y,shadow=True)
plt.legend(loc='lower right')
plt.show()
```
